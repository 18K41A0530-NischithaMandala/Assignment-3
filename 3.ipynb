```python
#3.Compute outer product of 2 vectors. 
#method 1
import numpy as np
p=[[1, 0], [0, 1]]
q=[[1, 2], [3, 4]]
np.outer(p,q)
```

```python
#method 2
p=[[1, 0], [0, 1]]
q=[[1, 2], [3, 4]]
x=[[],[],[],[]]
a=0
for i in p:
    for j in i:
        for k in q:
            for l in k:
                x[a].append(j*l)
        
        a+=1
        
for i in x:
    print(i)
        
        
```
